```python
import pandas as pd
import numpy as np
from math import isnan
```

# Champs Elysees

```python
#import csv file into dataframe
df = pd.read_csv("washington.csv", sep = ';')
df
```

```python
#split date & time, and get weekday
df['Date et heure de comptage'] = pd.to_datetime(df['Date et heure de comptage'], utc = True)
df['date'] = df['Date et heure de comptage'].dt.date
df['heure'] =df['Date et heure de comptage'].dt.hour
df['weekday'] = df['Date et heure de comptage'].dt.day_of_week


#filter by noeud amont and noeud aval
df = df[(df["Libelle noeud amont"] == "Av_Champs_Elysees-Washington") & (df["Libelle noeud aval"] == "Av_Champs_Elysees-Berri")]

#select useful cols
dfce = df[["date", "heure", "weekday", "Débit horaire", "Taux d'occupation"]]
dfce = dfce.sort_values(['date', 'heure'], ascending= [True, True])

dfce
```

```python
#checking for null values
dfce.isna().sum()
```

```python
#deal with nulls 
#replace null by mean value with the same 'weekday' & 'heure' values

for idx, row in dfce.iterrows():

  if not isnan(row[-2]) or not isnan(row[-1]):
    continue

  if isnan(row[-2]) and isnan(row[-1]):
    day = row['weekday']
    hour = row['heure']
    avg1 = dfce[(dfce['weekday'] == day) & (dfce['heure'] == hour)].mean()['Débit horaire']
    avg2 = dfce[(dfce['weekday'] == day) & (dfce['heure'] == hour)].mean()["Taux d'occupation"]
    dfce.at[idx, 'Débit horaire'] = avg1
    dfce.at[idx, "Taux d'occupation"] = avg2


print(dfce.isna().sum()) #check whether imputation has done

dfce = dfce.dropna(axis = 0, how = 'any') #drop columns that still have nulls 
dfce

```

# Convention

```python
#split date & time, and get weekday
df['Date et heure de comptage'] = pd.to_datetime(df['Date et heure de comptage'], utc = True)
df['date'] = df['Date et heure de comptage'].dt.date
df['heure'] =df['Date et heure de comptage'].dt.hour
df['weekday'] = df['Date et heure de comptage'].dt.day_of_week


#filter by noeud amont and noeud aval
df = df[(df["Libelle noeud amont"] == "Lecourbe-Convention") & (df["Libelle noeud aval"] == "Convention-Blomet")]

#select useful cols
dfc = df[["date", "heure", "weekday", "Débit horaire", "Taux d'occupation"]]
dfc = dfc.sort_values(['date', 'heure'], ascending= [True, True])

dfc
```

```python
#checking for null values
dfc.isna().sum()
```

```python
#deal with nulls in Débit horaire col
#replace null by mean value with the same 'weekday' & 'heure' values

for idx, row in dfc.iterrows():

  if not isnan(row[-2]):
    continue

  if isnan(row[-2]):
    day = row['weekday']
    hour = row['heure']
    avg = dfc[(dfc['weekday'] == day) & (dfc['heure'] == hour)].mean()['Débit horaire']
    dfc.at[idx, 'Débit horaire'] = avg


print(dfc.isna().sum()) #check whether imputation has done

dfc = dfc.dropna(axis = 0, how = 'any') #drop columns that still have nulls 
dfc
```
